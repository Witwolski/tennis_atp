```python
#model_file2
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

from sklearn.metrics import classification_report

# Step 2: Split the data into features (X) and target (y)
X = data.drop(columns=['Winner_IsFav'])
y = data['Winner_IsFav']

# Step 3: Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Step 4: Train a logistic regression model
model = LogisticRegression(max_iter=10000000)
model.fit(X_train, y_train)

# Step 5: Make predictions on the test set
y_pred = model.predict(X_test)

# Step 6: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))

# Get the coefficients or feature importances
feature_importances = model.coef_[0]  # For logistic regression, retrieve the coefficients
# Alternatively, for tree-based models like decision trees or random forests, use feature_importances_ attribute

# Rank the features based on importance
feature_importance_dict = dict(zip(X.columns, feature_importances))
sorted_features = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)
print("Feature Importance:")
feature_to_keep=[]
for feature, importance in sorted_features:
    if importance>0 or importance<-0:
        feature_to_keep.append(feature)
        #print(f"{feature}: {importance}")

from joblib import dump
dump(model, 'model_file2_all.joblib')
```
